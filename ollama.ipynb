```python
import base64
```

```python
# passing single image
image_path='Linkedin (1).jpg'
with open(image_path,'rb') as f:
    image_bytes=f.read()
image_64=base64.b64encode(image_bytes).decode('utf-8')
response=ollama.generate(model='gemma3:4b',images=[image_64],prompt='give caption for the image',stream=True)
```

```python
for i in response:
    print(i['response'],end='')
```

```python
# passing multiple images
image_paths=['Linkedin (1).jpg','Green AI (1).png']
images_base64=[]
for i in image_paths:
    with open(i,'rb') as f:
        image_bytes=f.read()
        images_base64.append(base64.b64encode(image_bytes).decode('utf-8'))

response=ollama.generate(model='gemma3:4b',images=images_base64,prompt='Generate an story based on these images, make sure you take context from each and every image.',stream=True)
```

```python
# response.response
for i in response:
    print(i,end='')
```

```python
# response.response
for i in response:
    print(i['response'],end='')
```
